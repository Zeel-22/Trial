```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
weather_data = pd.read_csv('/content/the weather of 187 countries in 2020.csv')
```

```python
columns_to_check = weather_data.columns
for col in columns_to_check:
    unique_values = weather_data[col].unique()
    if len(unique_values) <= 100:
        print(f"Unique values in column '{col}': {unique_values}")
        for value in unique_values:
            count = (weather_data[col] == value).sum()
            print(f"Count of '{value}': {count}")
```
